```python
# Convertir los datos de la columna 'genres' a cadenas
df_games_clear['genres'] = df_games_clear['genres'].astype(str)

# Función para eliminar "Free to Play" de los géneros
def eliminar_free_to_play(generos):
    try:
        lista_generos = ast.literal_eval(generos)  # Convertir cadena a lista
        # Eliminar "Free to Play" si está presente
        if "Free to Play" in lista_generos:
            lista_generos.remove("Free to Play")
        # Convertir de nuevo a cadena y devolver
        return ', '.join(lista_generos)
    except (SyntaxError, ValueError):
        return None

# Aplicar la función a la columna 'genres'
df_games_clear['genres'] = df_games_clear['genres'].apply(eliminar_free_to_play)
```

```python
# Dividir la cadena por comas y tomar el primer elemento
df_games_clear['genres'] = df_games_clear['genres'].str.split(',').str[0].str.strip()

#Mostrar el resultado
print(df_games_clear['genres'])
```

```python
# Define a function to handle price conversion
def handle_price(price):
    if isinstance(price, str):
        # Define a dictionary to map substrings to their replacements
        replacements = {
            'Free to Play': 'Free',
            'Play for Free': 'Free',
            'Free to Try': 'Free',
            'Play for Free!': 'Free',
            'Free to Use': 'Free',
            'Free Demo': 'Free'
        }
        # Replace substrings using the dictionary
        for old, new in replacements.items():
            price = price.replace(old, new)
        return price
    else:
        return price  # Leave non-string values unchanged

# Function to convert non-'Free' strings to NaN
def convert_to_nan(value):
    if isinstance(value, str) and value != 'Free':
        return np.nan
    else:
        return value

# Apply the handle_price function to the 'price' column
df_games_clear['price'] = df_games_clear['price'].apply(handle_price)

# Apply the convert_to_nan function to convert non-'Free' strings to NaN
df_games_clear['price'] = df_games_clear['price'].apply(convert_to_nan)
```
