Importar Bibliotecas Necesarias:

```python
import pandas as pd 
import json as js 
import ast as ast
import gzip
from pandas import json_normalize
```

Ruta del Archivo e Inicialización:

```python
path_reviews = r'C:\Users\Owner\Desktop\LABS\Proyecto_STEAM\datos_STEAM\user_reviews.json.gz'
```

```python
lista_reviews = []
with gzip.open(path_reviews, 'rt', encoding='utf-8') as file:
    for line in file.readlines():
        lista_reviews.append(ast.literal_eval(line))

df_reviews = pd.DataFrame(lista_reviews)
df_reviews_clean = df_reviews.copy()
```

```python
df_reviews_clean
```

needed to import json normalize here

```python
df_expanded = df_reviews_clean.explode('reviews')
```

```python
df_expanded.reset_index(drop=True, inplace=True)
```

```python
df_reviews_normalized = json_normalize(df_expanded['reviews'])
```

```python
df_final = pd.concat([df_expanded.drop(columns=['reviews']), df_reviews_normalized], axis=1)
```

```python
columns_to_drop = ['funny', 'last_edited', 'helpful']
df_reviews_clean = df_final.drop(columns=columns_to_drop, errors='ignore')

df_reviews_clean
```

```python
df_reviews_clean.to_csv('reviews_clean.csv', index=False)
```

```python
df_reviews_clean.to_parquet('reviews_clean.parquet', index=False)
```
